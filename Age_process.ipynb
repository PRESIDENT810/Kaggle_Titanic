```python
import math

def append_age(row,axis=0):    
    global data_withoutAge
    global data_withAge
    global data_MrAge
    global data_MrsAge
    global data_MissAge
    global data_MasterAge
    global data_DrAge
    global data_OtherAge
    
    i = row["PassengerId"]-1
    if math.isnan(row["Age"]):
        data_withoutAge = pd.concat([data_withoutAge,data[i:i+1]],axis=0)
    else:
        data_withAge = pd.concat([data_withAge,data[i:i+1]],axis=0)
        name = row["Name"]
        if "Mrs" in name:
                data_MrsAge = pd.concat([data_MrsAge,data[i:i+1]],axis=0)
        elif "Mr" in name:
                data_MrAge = pd.concat([data_MrAge,data[i:i+1]],axis=0)
        elif "Miss" in name:
                data_MissAge = pd.concat([data_MissAge,data[i:i+1]],axis=0)
        elif "Master" in name:
                data_MasterAge = pd.concat([data_MasterAge,data[i:i+1]],axis=0)
        elif "Dr" in name:
                data_DrAge = pd.concat([data_DrAge,data[i:i+1]],axis=0)
        else:
            data_OtherAge = pd.concat([data_OtherAge,data[i:i+1]],axis=0)
    return
```

```python
def sep_age():
    data.apply(append_age,axis=1)
age_data_list = sep_age()
```

```python
def fill_age(row):
    name = row["Name"]
    passengerid = row["PassengerId"]
    if "Mrs" in name:
        age = np.random.normal(Mrs_mean,Mrs_std)
    elif "Mr" in name:
        age = np.random.normal(Mr_mean,Mr_std)
    elif "Miss" in name:
        age = np.random.normal(Miss_mean,Miss_std)
    elif "Master" in name:
        age = np.random.normal(Master_mean,Master_std)
    elif "Dr" in name:
        age = np.random.normal(Dr_mean,Dr_std)
    else:
        age = np.random.normal(Other_mean,Other_std)
    row["Age"] = round(age,1)
    return row
```

## apply 

```python
def data_fillAge(data_withoutAge):
    data_withoutAge = data_withoutAge.apply(fill_age,axis=1)
    return data_withoutAge
data_AgeFilled = data_fillAge(data_withoutAge)
```
